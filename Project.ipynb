```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset from CSV file
heart_df1=pd.read_csv('/content/heart_2020_cleaned.csv').head(1000)
heart_df1
```

```python
# Load the dataset from CSV file
heart_df2=pd.read_csv('/content/heart.csv').head(1000)
heart_df2
```

```python
# Add ID column datasets
heart_df1['Person_ID']=range(1, len(heart_df1) + 1)
heart_df1
```

```python
# Add ID column datasets
heart_df2['Person_ID']=range(1, len(heart_df2) + 1)
heart_df2
```

```python
# Check the data types, non-null counts, and memory usage
heart_df1.info()
```

```python
heart_df2.info()
```

```python
# Replace AgeCategory
age_mapping={
    '18-24': 18, '25-29': 25, '30-34': 30, '35-39': 35,
    '40-44': 40, '45-49': 45, '50-54': 50, '55-59': 55,
    '60-64': 60, '65-69': 65, '70-74': 70, '75-79': 75,
    '80 or older': 80
}
heart_df1['Age']=heart_df1['AgeCategory'].apply(lambda x: age_mapping[x])
heart_df1
```

```python
# Convert categorical variables to numeric format for analysis
heart_df1['Sex']=heart_df1['Sex'].replace({'Male': 1, 'Female': 0}).astype(int)
heart_df1['HeartDisease']=heart_df1['HeartDisease'].replace({'Yes': 1, 'No': 0}).astype(int)
binary_cols=['Smoking','AlcoholDrinking','Stroke','DiffWalking','PhysicalActivity','Asthma','KidneyDisease','SkinCancer']
heart_df1[binary_cols]=heart_df1[binary_cols].replace({'Yes': 1, 'No': 0}).astype(int)
heart_df1['Diabetic']=heart_df1['Diabetic'].replace({'No': 0,'No, borderline diabetes': 1,'Yes (during pregnancy)': 2,'Yes': 3})
heart_df1
```

```python
# Combine two datasets on 'Person_ID' to create a single dataframe
merged_df=pd.merge(heart_df1, heart_df2,on='Person_ID', how='inner')
merged_df
```

```python
# Remove duplicate or redundant columns after merge
merged_df=merged_df.drop(columns=['age','sex'])
merged_df
```

```python
# Check the data types, non-null counts, and memory usage
merged_df.info()
```

```python
# Check for missing values in the dataset
merged_df.isnull().sum()
```

```python
# Check for duplicate rows in the dataset
merged_df.duplicated().sum()
```

```python
# Select Clinical features
clinical_features=['Age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'target']
clinical_corr=merged_df[clinical_features].copy().corr()
clinical_corr
```

```python
# Visualize correlation among clinical features
plt.figure(figsize=(7,6))
sns.heatmap(clinical_corr, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.title('Clinical Feature Correlation Heatmap')
plt.show()
```

```python
# Select Behavioural features
behavioral_features=['BMI', 'Smoking', 'PhysicalActivity', 'SleepTime', 'HeartDisease']
behavioral_corr=merged_df[behavioral_features].copy().corr()
behavioral_corr
```

```python
# Visualize correlation among behavioral features
plt.figure(figsize=(6,5))
sns.heatmap(behavioral_corr, annot=True, fmt=".2f", cmap='viridis', linewidths=0.5, linecolor='black')
plt.title('Behavioral Feature Correlation Heatmap')
plt.show()
```

```python
# Calculate Average Heart Disease Rate for each Sleep Duration
sleep_heart=merged_df.groupby('SleepTime')['HeartDisease'].mean().reset_index()
sleep_heart
```

```python
# Plotting Sleep Time vs Heart Disease
plt.figure(figsize=(8,5))
plt.bar(sleep_heart['SleepTime'], sleep_heart['HeartDisease'], color='skyblue', edgecolor='black')
plt.title('Sleep Time vs Heart Disease')
plt.xlabel('Sleep Time (hours)')
plt.ylabel('Average Heart Disease Rate')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calculate Average Physical Activity
pa_genhealth=merged_df.groupby('GenHealth')['PhysicalActivity'].mean()
pa_genhealth
```

```python
# Plotting Physical Activity by General Health
pa_genhealth.plot(kind='bar', color='mediumseagreen')
plt.title('Physical Activity by General Health')
plt.xlabel('General Health')
plt.ylabel('Average Physical Activity')
plt.show()
```

```python
# Make a copy
df=merged_df.copy()
# Define BMI bins and categories
bins=[0, 18.5, 25, 30, 35, 40, 50]
labels=['Underweight','Normal','Overweight','Obese I','Obese II','Obese III']
# Create BMI category column
df.loc[:, 'BMI_category'] = pd.cut(df['BMI'], bins=bins, labels=labels, right=False)
# Count patients in each BMI category by heart disease status
bmi_counts=df.groupby(['BMI_category', 'HeartDisease'], observed=False).size().unstack(fill_value=0)
# Plotting as grouped bar chart
bmi_counts.plot(kind='bar', figsize=(8,5), color=['skyblue','salmon'], alpha=0.7)
plt.title('BMI Distribution by Heart Disease Status')
plt.xlabel('BMI Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(['No Heart Disease', 'Heart Disease'])
plt.show()
```

```python
# Create readable labels for HeartDisease column
merged_df.loc[:, 'HeartDisease_Text']=merged_df['HeartDisease'].replace({0:'No Disease', 1:'Disease'})
# Count patients by chest pain type and heart disease status
chest_pain_counts=merged_df.groupby(['cp','HeartDisease_Text']).size().reset_index(name='Count')
plt.figure(figsize=(10,6))
sns.barplot(
    x='cp',
    y='Count',
    hue='HeartDisease_Text',
    data=chest_pain_counts,
    palette={'No Disease':'#34495e', 'Disease':'#f39c12'}
)
# Plotting Patient Count by Chest Pain Type and Disease Status
plt.title('Patient Count by Chest Pain Type and Disease Status')
plt.xlabel('Chest Pain Type')
plt.ylabel('Patient Count')
plt.legend(title='Heart Disease Status')
plt.show()
```
